# Pair Plot

```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Pair Plot : a data visualization tool that displays pairwise relationships between variables in a dataset.

# it is used in machine learning for comparision different graphs in a single a graph.

df = sns.load_dataset("tips").head(100) # type: ignore
df
```

# Parameter


```python
# vars : for choosing parameters , which we want to plot  only.
# hue_order :  hue_order parameter in Seaborn allows for the explicit control of the order.

# palette : for color combinations
# x_vars  : for changing the varibles according to the axis
# kind  : "scattere","kde","hist","reg" for best visuilization of graph as per the need.

# diag_kind : controls the type of plot displayed on the diagonal subplots.

# marker : for changing the marker.

sns.pairplot(df,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],palette="PuBu",x_vars=["total_bill","tip"],y_vars=["tip","size"],kind="kde",diag_kind="hist",markers={"*",">"})

plt.show()
```
