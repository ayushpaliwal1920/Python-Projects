# HeatMap

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# HeatMap => represent two-dimensional data where individual values within a matrix are depicted by colors. This allows for quick identification of patterns, correlations, and trends within a dataset.

#  sns.heatmap(data = data) then plt.show()

var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var)
```

```python
#  when we are using data from csv. : we cnnot make the heatmap of the column consisting String data

df = sns.load_dataset("anagrams").head(10)
x = df.drop(columns=["attnr"],axis=1)
x
```

# Parameters.

```python
# vmin and vmax : for changing the  minimum and maximum values in the color bar
# cmap : for color combinations.
# annot : for showing the values which is on the data.
# fmt : format of data "s" for strings.
# annot_kws : {" " : " "} : when we want to change the multiple characterstics.pass dict 
# linewidth : for making the gap btwn boxes.
# linecolor : for changing the color of background
# cbar : passing False for removing the colorbar.
# xticklabels : for removing labels of x axis pass False.
# yticklabels : for removing labels of y axis pass False.
# set() : uses for setting labels in seaborn .



y = {"fontsize":10,"color":"g"}
v = sns.heatmap(x,vmin=0,vmax=12,cmap="PuBu",annot=True,annot_kws=y,linewidths=2,linecolor="b",cbar=False,xticklabels=False)

v.set(xlabel = "Python",ylabel="Ayush")
sns.set(font_scale=1)
plt.show()
```

```python
var1 = np.linspace(1,10,10).reshape(2,5)
var1
```

```python
# if we want to change the number or data inside  the boxes .
# annot & fmt  : fmt is format , which of data is passing.and annot what data we are passing ex is below

sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',annot=True)
```

```python
arr = np.array([["a0","a1","a2","a3","a4"],["b0","b1","b2","b3","b4"]])
arr
```

```python
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',annot=arr,fmt="s")
```
