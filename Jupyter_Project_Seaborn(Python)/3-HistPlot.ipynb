# Histplot in seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
# histogram plot : in seaborn it use ,
# df = sns.load_dataset("") then sns. distplot(df[""]) then  plt.show( )

df = sns.load_dataset("penguins").head(50)
df
```

<!-- # Paramters -->

```python
# bins : refer to the intervals or ranges used to group continuous data when creating a histogram. 

# kde :Kernel Density Estimate plot,it is line which shows actual plot line .

# rug : we can show krnal density at the x axis
# color : colors

# logscale : uses for changing graph into logarithmic graph


sns.displot(df["flipper_length_mm"],bins =[170,180,190,200,210,220,230],kde=True,rug=True,color="g",log_scale=True)
plt.show()
```
