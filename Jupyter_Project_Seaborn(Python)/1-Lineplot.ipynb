# Line plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Line plot =  data=sns.load_dataset(" ")
#              sns.lineplot(x=" ",y=" ",data=data)

var = [1,2,3,4,5,6,7]
var_1 =[2,3,4,15,6,5,4]

x_1 = pd.DataFrame({"var":var,"var_1":var_1})
sns.lineplot(x="var",y="var_1",data=x_1)

# we can also plot graph without using dataframe : 
# sns.lineplot(x=var,y=var_1)

plt.show()
```

```python
# load_dataset : we can use this function to load datasets from a external file to make a plot

y_data = sns.load_dataset("titanic").head(100)
y_data
```

# Parametes

```python
# hue :   It enables the comparison of different groups within the same plot
# size :  customize size of the graph 
# style : change the style of lines
# palette : The palette parameter in Seaborn is used to control the color scheme of your plots

# marker : uses for maker in the peaks.
# dashes : it primarily refers to controlling the linestyle of lines in plots
# legend : in seaborn it is used to remove label.

sns.lineplot(x="age",y="fare",data=y_data,hue="sex",size=50,style="sex",palette="Accent",markers=["o","*"],dashes=False,legend=False)

plt.grid()
plt.show()
```
