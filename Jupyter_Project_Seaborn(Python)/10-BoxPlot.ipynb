# Box Plot

```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Box plot : boxplot() function creates box plots, also known as box-and-whisker plots.

df = sns.load_dataset("tips")
df
```

# Parameters

```python
#  order : change order of labels in axises.
#  showmeans : for showing the mean of of data.
#  meanprops : for changing the properties of the mean.passes dict as argument
# linewidth : for increasing and decreasing line width fo boxes.
# linecolor : for changing the color of the line of boxes.
# orient : for changing orientation but data on both axies should be numerical data or y axis always contains numerical data.

# 

sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="time",palette="rocket_r",order=["Fri","Sat","Sun","Thur"],showmeans=True,meanprops={"marker":"+","markeredgecolor":"w"},linewidth=2,linecolor="b",orient="v")
plt.show()
```

# Single box blot

```python
sns.set(style="whitegrid")
sns.boxplot(x=df["total_bill"])
plt.show()
```
