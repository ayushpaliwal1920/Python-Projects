```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# graph = sns.FacetGrid(df,col="",hue="") then graph.map(plt.scatter).add_legend() then plt.show()

df = sns.load_dataset("tips").head(100)
df
```

```python
# for making mulple graph : gvie column parameter as differentiating parameter.plt 


fg = sns.FacetGrid(df,col="day",hue="sex")

fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```
