# Count Plot : 
 
#Count plot plotts the count of the numbers of records by category.
#Barplot plots a value or metric for each category(by default,barplot plots the mean of a varible,by category )

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Count plot : df = load_dataset("") then sns.countplot(x="",data=df : it is for overall counting.

df = sns.load_dataset("tips").head(100)
df
```

# Parameters

```python
# for horizotal : give y parameter.
# palette : for xolor combination.
# color : for colors
# saturation : for color saturations

sns.countplot(y ="sex",data=df,hue="day",palette="crest",saturation=0.2)
plt.show()
```
