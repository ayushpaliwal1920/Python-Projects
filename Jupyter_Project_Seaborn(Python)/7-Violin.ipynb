Violin


```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# A violin plot plays a similar role as a box and whisker plot.it shows the distribution of quantitative data across several levels of ine (or more) categorical variables such that those distributions canbe comapared.

# A violin plot in Seaborn is a data visualization that displays the distribution of quantitative data across different levels of one or more categorical variables. It combines the features of a box plot with a kernel density estimate (KDE) of the data distribution.

# sns.set(style ='whitegrid) then f = sns.load_dataset("fmri") then sns.violinplot(x="",y="",data=f)
 
df = sns.load_dataset("tips").head(100)
df
```

# Parameters

```python
# hue : uses for comparing two datas.
# linewidth : for changing the width of the line.
# linecolor : changing the color of line
# palette   : for combination of colors.
# order     : for changing the order.The order parameter in Seaborn's violinplot function controls the order in which categorical groups appear on the plot's categorical axis.

# saturation : for opacity or color saturation.
# color  : for changing the color for all diagram.
# split  : the split parameter is a boolean argument that controls whether the violins should be split to show the distributions of two different levels of a hue variable within the same violin.

# scale :the scale parameter controls how the width of each violin is determined.it passes parameter "area","count","width"

# for changing orientation : interchange the x and y parameter.

# inner : violinplot function controls what is displayed inside the violins, providing additional information about the data distribution,we can pass "box","quartile","point","stick",None.


sns.set(style='whitegrid')
sns.violinplot(x="day",y="total_bill",data=df,hue="time",linewidth=2,palette="ocean",linecolor="w",saturation=0.2)
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=df,hue="day",order=["Dinner","Lunch"],split=True,inner="stick")

plt.show()
```

# Single voilin Plot

```python
sns.violinplot(x=df["total_bill"])
plt.show()
```

```python
sns.violinplot(y=df["total_bill"])
plt.show()
```
