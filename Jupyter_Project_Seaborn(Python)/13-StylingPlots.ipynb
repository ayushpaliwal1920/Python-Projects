# Styling Plots


```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Seaborn Figures Styles , Removing Axes Spines , Scale and Context.

df = sns.load_dataset("tips").head(100)
df
```

# Parameters and fxns

```python
# set_style() fxn : uses various parameters and styles.
# despine() fxn : for removing axes.
# figure() fxn : for changing figure size
# set_context fxn : for giving content,changing scaling.

sns.set_style("darkgrid")

sns.set_context("paper",font_scale=0.8)
plt.figure(figsize=(5,10))
sns.barplot(x="day",y="total_bill",data=df,hue="sex",palette="summer")
sns.despine()
plt.show()
```
