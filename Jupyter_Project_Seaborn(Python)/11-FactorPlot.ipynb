# Factor Plot

```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Factor Plot : seaborn.factorplot() method is used to draw a categorical plot onto a FacetGrid

df = sns.load_dataset("tips")
df
```

```python
sns.catplot(x="size",y="tip",data=df,hue="sex",palette="PuBu")
plt.show()
```
