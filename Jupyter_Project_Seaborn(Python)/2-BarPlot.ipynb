# Barplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# BarPlot : df = sns.load_dataset(' ') then sns.barplot(x=" ",y=" ",data =df) then plt.show()

x=[1,2,3,4,5,6,7]
y=[20,30,40,50,60,20,10]

df = sns.load_dataset("penguins").head(50)
df
```

```python
# sns.barplot(x="island",y="bill_length_mm",data=df)

sns.barplot(x=df.island,y=df.body_mass_g,data=df,hue="sex")
plt.show()
```

# Parameter

```python
# order : we can change order of the graph
# hue_order : used to change order of bars
# ci : sed to control the estimation and visualization of confidence intervals around statistical estimates in plots

# n_boot : manages size
# orientat : horizontal when we have  x and y both numerical parameter : "h"
# color : for colors

order_1 = ["Dream","Torgersen","Biscoe"]

sns.set(style="darkgrid")
sns.barplot(x=df.island,y=df.body_mass_g,data=df,hue="sex",order=order_1,hue_order=["Female","Male"],n_boot=2,orient="v",color="g")

plt.show()
```

```python
# palette : for color combinations
# saturation : for changing hardness of colors
# errcolor :  allows customization of the color of error bars in plots that display them
# errwidth : allows customization of the width of error bars in plots that display them

# capsize : uses for cap on the top of the bars
# dodge : for overlapping both bars , use True or false for see deference.

# alpha : uses for opacity 
# 

order_1 = ["Dream","Torgersen","Biscoe"]

sns.set(style="darkgrid")

sns.barplot(x=df.flipper_length_mm		,y=df.body_mass_g,data=df,orient="h",palette="RdBu",saturation=0.5,errcolor="m",errwidth=2,capsize=0.3,dodge=False)

plt.show()
```
