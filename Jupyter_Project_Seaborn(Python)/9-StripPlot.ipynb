# Strip Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
# Strip Plot : A strip plot is very simple to understand .It is basicallly a scatter plot that differentiates different categories .Srip plots are considered a good altenative to a box plot or a violin plot.

df = sns.load_dataset("tips").head(100)
df
```

# Parameters

```python
# linewidth : for changing the size of outline.
# edgecolor : changing the color of edges of dots/marker.

# jitter : for changing the displacement btw data   plots.
# size  : for changing the size of dots.
# marker : choose marker.
# alpha  : for opacity.

# interchage the x and y for vertical and horizontal plot.


sns.stripplot(x="day",y="total_bill",data=df,hue="sex",palette="rocket_r",linewidth=0.5,edgecolor="m",jitter=0.2,size=10,marker="*",alpha=0.6)
plt.show()
```
