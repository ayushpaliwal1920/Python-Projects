# Stem Plot

```python
import matplotlib.pyplot as plt
```

```python
# Stem plot : uses plt.stem(x,y) method to make a chart
#  it is a dot and linear plot.

x = [1,2,3,4,5,6]
y = [2,4,5,2,4,5]

plt.stem(x,y)
plt.show()
```

# Parameters :


```python
# Linefmt : uses for changing the lines of the plot 
        #  "'-'" - solid lines
        #  "'--'" - Dashed lines
        #  "'-.'" - Dash-dot lines
        #  "':'" - Dotted lines

# markerfmt : uses for changing the shape of peak or corners.

# bottom : uses for changing the bottom line.
# basefmt : for changing the base line color.
# label : uses for giving the label .for showing the label we need legend fxn .
# use_line_collection : uses for the changing the color of lines as the base line. but it is removed in new versions.

# orientation : changing the orientation v or h 

x = [1,2,3,4,5,6]
y = [2,4,5,2,4,5]

plt.stem(x,y,linefmt="--",markerfmt="r"+"*",bottom=2,basefmt="g",label="Python",orientation="horizontal")
plt.legend(loc=1)
plt.show()
```
