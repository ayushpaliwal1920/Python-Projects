## Parameters

```python
# Historgram Plot : uses  hist() method .
# Parameters :

# color = uses for color 
# edgecolor = uses for edges
# bins  = this parameter determines how data is grouped into intervals ,uses a list of number from which we want to make  intervals ex- l=[10,20,30,40,50,60] or we can pass "auto" to the bins parameter

# we can change the range : using (r1,r2) 

# cumulative = this parameter reverses the frequency bar graph when we give the value -1

# bottom = this parameter controls the baseline of the bars. means if we want to start graph from 10 ,20 or etc then pass value 10,20 etc

# align = alignment of the graph

# histtype = parameter uses for changing the histogram type ex step,bar,stepfilled etc

# orientation = convert the orientation vertical or horizontal

# rwidth = uses for manuplation of width btw intervals

# log = "true " uses for converting table into log table

# label = for giving label followed by legend() method



# axvline() : this method is used for seperating data with a line , and uses three parameters , axvline(val,color="-",label="-")


# grid() : this method is for showing grid lines

numX = [53, 47, 49, 45, 14, 36 ,46, 23, 11, 50, 52, 37 ,47 ,34 ,18, 19, 48, 44 ,59 ,21 ,13 ,59 ,25 ,22,
 21, 41, 12 ,45 ,42 ,25 ,59 ,34 ,36, 16, 59, 46, 18 ,44 ,36 ,40 ,12 ,58 ,43 ,58 ,53, 16, 47, 17,
 55, 22]

l=[10,20,30,40,50,60]

# plt.hist(numX,color="g",bins="auto",edgecolor = "w")

# plt.hist(numX,"auto",(10,100),edgecolor = "w") ## range

plt.hist(numX,color="g",bins=l,edgecolor = "r",cumulative = -1,bottom=10,histtype="stepfilled",rwidth=0.8,log="True",label="Ayush")

plt.legend()
plt.title("Demo Histogram")
plt.xlabel("Python")
plt.ylabel("Time Complexity")

# axvline()
plt.axvline(30,color="b",label="line")
plt.grid()
plt.show()
```
