```python
import matplotlib.pyplot as plt
```

```python
# 'Line' Plot : use fxn "plt.bar(x,y) & plt.show()"

x = ["ayush" , "python" , "c++" , "clanguage","java"]
y = [85,90,55,65,55]

plt.plot(x,y)  # for bar graph
plt.xlabel("languages",fontsize=40)  # x axis representation name 
plt.ylabel("popularity percentage",fontsize=20) # y axis representation name
plt.title("FOT student's data")
plt.show()

```

## Parameters


```python
import matplotlib.pyplot as plt

# Data
x = ["ayush", "python", "c++", "clanguage", "java"]
y = [85, 90, 55, 65, 55]

# Line chart parameters
plt.plot(
    x, 
    y, 
    color="b",          # line color
    linestyle=":",      # line style
    linewidth=3,        # line thickness
    marker="o",         # marker for each point
    markerfacecolor="y",# marker fill color
    markeredgecolor="r",# marker border color
    alpha=0.8,          # line opacity
    label="Percentage"  # label for legend
)

# Add legend and labels
plt.legend()
plt.xlabel("Languages", fontsize=20)
plt.ylabel("Popularity Percentage", fontsize=20)
plt.title("FOT Student's Data")

# Show the chart
plt.show()
```

## Multiple BarGraph

```python
import matplotlib.pyplot as plt

# Data
x = ["ayush", "python", "c++", "clanguage", "java"]
y = [85, 90, 55, 65, 55]
z = [10, 20, 30, 40, 50]

# Plot first line
plt.plot(
    x, y,
    color="b",
    linestyle="-",
    marker="o",
    linewidth=2,
    label="Dancing Performance"
)

# Plot second line
plt.plot(
    x, z,
    color="r",
    linestyle="--",
    marker="s",
    linewidth=2,
    label="Singing Performance"
)

# Labels and title
plt.xlabel("Languages", fontsize=10)
plt.ylabel("Popularity Percentage", fontsize=10)
plt.title("FOT Student's Data")

# Legend (correct usage)
plt.legend(loc="center")  # you had 'centre' (wrong spelling)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = ["ayush", "python", "c++", "clanguage", "java"]
y = [85, 90, 55, 65, 55]
z = [10, 20, 30, 40, 50]

# Plot first line (Dancing Performance)
plt.plot(
    x, y,
    color="b",
    marker="o",
    linestyle="-",
    linewidth=2,
    label="Dancing Performance"
)

# Plot second line (Singing Performance)
plt.plot(
    x, z,
    color="r",
    marker="s",
    linestyle="--",
    linewidth=2,
    label="Singing Performance"
)

# Labels and title
plt.xlabel("Languages", fontsize=10)
plt.ylabel("Popularity Percentage", fontsize=10)
plt.title("FOT Student's Data")

# Add legend
plt.legend()

# Optional: rotate x labels slightly for readability
plt.xticks(rotation=10)

# Show the plot
plt.show()
```

# Horizontal BarGraph


```python
import matplotlib.pyplot as plt

# Data
x = ["ayush", "python", "c++", "clanguage", "java"]
y = [85, 90, 55, 65, 55]
z = [10, 20, 30, 40, 50]

# Plot first line (Dancing Performance)
plt.plot(
    x, y,
    color="b",
    linestyle="-",
    marker="o",
    linewidth=2,
    label="Dancing Performance"
)

# Plot second line (Singing Performance)
plt.plot(
    x, z,
    color="r",
    linestyle="--",
    marker="s",
    linewidth=2,
    label="Singing Performance"
)

# Labels and title
plt.xlabel("Languages", fontsize=10)
plt.ylabel("Popularity Percentage", fontsize=10)
plt.title("FOT Student's Data")

# Correct legend placement
plt.legend(loc="center")

# Rotate x-axis labels slightly for readability
plt.xticks(rotation=10)

# Display the chart
plt.show()
```
