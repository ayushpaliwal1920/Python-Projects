## Pie Plot


```python
import matplotlib.pyplot as plt

# IMP 

x = [10,20,30,40]
y = ["c","c++","java","Pyhon"]
```

```python
# Pie Plot : uses pie method and show fxn for showing pie chart
# pie() : pie function is used for ploting pie chart , followed by plt.show() fxn

plt.pie(x)
plt.show()

```

# Parameters

```python
# labels : use two parameters for labels, make a list of var which we want to make label, as we did making of list y

# explode : he explore parameter enables interactive data exploration directly in your plots means if we want to show a part of piechart distinctvely

# colors : uses for giving color to the divisions

# autopct : uses for showing the percentage of the divisions , but values pases as ex  :  autopct="%0.3f%%"

# shadow : this is used for 3d look
# radius : for changing the radius of the piechart
# labeldistance : uses for changing the distance of labels.
# startangle : uses for changing/rotating the angle of the circle / chart.
# textprops : data pases as a dict, uses for changing the font size or other properties of the labels.

# counterclock : uses for changing the dirxn of the diagram into clock wise dirxn use counterclock = Flase

# wedgeprops : uses for changing the edges , passes dict as parameters.
# centre : uses for setting the centre.
# rotatelabel : uses for rotating the labels
# legend() fxn : uses for setting labels as titles

ex = [0.0,0.3,0.0,0.0]
clr = ["r","b","g","y"]

plt.pie(x , labels=y,explode=ex,colors=clr, autopct="%0.1f%%",shadow=True,radius=1,labeldistance=1.2,startangle=90,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":4,"edgecolor":"w"},rotatelabels=True)

plt.title("Clg chart")
plt.legend(loc = 3)

plt.show()
```

# DoNut Pie chart or multiple pie chart

```python
x = [10,20,30,40]
y = ["c","c++","java","Pyhon"]
x1 = [40,30,20,50]


c = ["r","b","g","y"]
plt.pie(x , labels=y,radius=1.5)
plt.pie(x,radius=0.5,colors=c)
plt.pie([1],colors="w") # if we want donut like shape white space of radius 0.5




plt.show()
```
