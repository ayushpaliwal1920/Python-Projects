# Step Plot

```python
import matplotlib.pyplot as plt
```

```python
# Step plot : plt.step(x,y)

x =[10,20,30,40,40,60,70]
y =[1,1,3,4,8,6,7]

plt.step(x,y,marker="o",ms="10",mfc="g",label="Python")
plt.title("Language")
plt.xlabel("people")
plt.ylabel("Languages")
plt.legend(loc=2)
plt.grid()
plt.show()
```
