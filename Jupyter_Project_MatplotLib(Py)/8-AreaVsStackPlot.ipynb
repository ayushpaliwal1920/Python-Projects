# Area or Stack Plot

```python
import matplotlib.pyplot as plt
```

```python
# Area or Stack plot : uses for showing multiple plot in a single diagram .
#                     uses plt.stackplot(x,y) minimum 2 paramters .

x=[10,20,30,50,10,60,90]
area=[1,2,3,4,5,6,7]

area1 = [1,101,20,50,520,5,11]
area2 = [13,52,55,66,10,55,00]
area3 = [14,55,20,78,95,66,33]

# Parameteres =>

# label = for labels,followed by legend.
# colors = for colors 
# baseline = for changing baseline . it has 3 values
    #      zero : no defference 
    #      sym  : for symmetry
    #      wiggle : check google

l=["area","area1","area2","area3"]
plt.stackplot(x,area,area1,area2,area3,labels=l,colors=["r","m","b"],baseline="wiggle")

plt.title("Python")
plt.legend(loc=1)
plt.show()
```
