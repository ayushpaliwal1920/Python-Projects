# Working with text


```python
import matplotlib.pyplot as plt
```

```python
# text = add text at an arbitrary location of the axis
# annotate = add an annotation ,with an optional arrow , at an arbitrary location of the axes.
# xlabel = add label
# ylabel = add label
# title = titles to the axes.

x =[1,2,3,4,5]
y=[3,5,2,6,8]

plt.plot(x,y)

plt.title("Py growth",fontsize=15)
plt.xlabel("days",fontsize=15)
plt.ylabel("Growth",fontsize=15)

# plt.text(2,5,"java",fontsize=15,style="italic",bbox={"facecolor":"r"})

# plt.annotate("c++",xy=(3,2),xytext=(4,4),arrowprops=dict(facecolor="black",shrink=100))

plt.legend(["up"],loc=0,shadow=True)
plt.show()
```
