## Matplotlib Bar Plot


```python
import matplotlib.pyplot as plt
```

```python
# Bar Plot : use fxn "plt.bar(x,y) & plt.show()"

x = ["ayush" , "python" , "c++" , "clanguage","java"]
y = [85,90,55,65,55]

plt.bar(x,y)  # for bar graph
plt.xlabel("languages",fontsize=40)  # x axis representation name 
plt.ylabel("popularity percentage",fontsize=20) # y axis representation name
plt.title("FOT student's data")
plt.show()

```

## Parameters


```python
# Parameters :

x = ["ayush" , "python" , "c++" , "clanguage","java"]
y = [85,90,55,65,55]

# Width :- uses '"width"' parameter for resizing the width of the bars
# Color :- uses '"color"' parameter for changing the parameter of the bars
# Multiple Color :- for multiple color we can make a list of different color and pass to the parameter.
# Align :- we can align x or y values using "align" parameter
# Boundary color / edgecolor : using "edgecolor" parameter we can change boundary.
# EdgeWidth/linewidth : using linewidth parameter we can adjust the size of boundary or outline.
# linestyle :- "linestyle" parameter for changing the style of the edge
# alpha parameter :- for adjusting the opacity values varies from 0 to 1.
# lable paramter with legend() method :- uses for setting label inside the bargraph uses label parameter followed by legend() fxn

colorB = ["y","b","g","m","r"]
# plt.bar(x,y , width=0.4 , color="y") 
plt.bar(x,y , width=0.4 , color=colorB,align="center",edgecolor = "r",linewidth=5,linestyle = ":",alpha =0.5, label="Percentage") 
plt.legend()
plt.xlabel("languages",fontsize=20)  # x axis representation name 
plt.ylabel("popularity percentage",fontsize=20) # y axis representation name
plt.title("FOT student's data")
plt.show()

```

## Multiple BarGraph


```python
#  We can bake multiple bargraph inside a chart. 
# 

# Overlapped bargraph ::

x = ["ayush" , "python" , "c++" , "clanguage","java"]
y = [85,90,55,65,55]
z = [10,20,30,40,50]

plt.bar(x,y,width =0.4,label="Dancing Performance",)
plt.bar(x,z,width =0.4,label="Singing Performance",color="r") # for another bar graph
plt.xlabel("languages",fontsize=10) 
plt.ylabel("popularity percentage",fontsize=10) 
plt.title("FOT student's data")
plt.legend('centre')
plt.show()
```

```python
# in a single graph multiple bargraphs:  

import numpy as np 

x = ["ayush" , "python" , "c++" , "clanguage","java"]
y = [85,90,55,65,55]
z = [10,20,30,40,50]
width =0.4 # decleared outside because its easy for many graphs 
p = np.arange(len(x)) # makes a array of indices of list x
print("list of indices :" , p) # just for understanding

p1 =[j+width for j in p] # making bargraph just ahead of previous one



plt.bar(p,y,width,label="Dancing Performance",)
plt.bar(p1,z,width,label="Singing Performance",color="r") # for another bar graph


plt.xlabel("languages",fontsize=10) 
plt.ylabel("popularity percentage",fontsize=10) 
plt.title("FOT student's data")

# giving name to bargraph elements : 
plt.xticks(p+width/2,x,rotation =10)
plt.legend()
plt.show()
```

# Horizontal BarGraph


```python
# use barh() method for horizontal bargraphs : barh(x,y) 

x = ["ayush" , "python" , "c++" , "clanguage","java"]
y = [85,90,55,65,55]
z = [10,20,30,40,50]

plt.barh(x,y,label="Dancing Performance",)
plt.barh(x,z,label="Singing Performance",color="r") # for another bar graph
plt.xlabel("languages",fontsize=10) 
plt.ylabel("popularity percentage",fontsize=10) 
plt.title("FOT student's data")
plt.legend('centre')
plt.show()
```
