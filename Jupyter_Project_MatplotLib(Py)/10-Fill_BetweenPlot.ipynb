# Fill Between plot


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Filbetween plot = it is used for filling the area btw two curves.
#                  plt.plot(x,y) then plt.fill_betweeen() then plt.show()


x=np.array([1,45,6,8,29,5,56,2,65])
y=np.array([1,22,3,44,5,6,7,81,90])

plt.plot(x,y,color="r")

plt.fill_between(x,y,color="g",where=(x>=29) & (x<=44),alpha =0.5)

plt.show()
```
