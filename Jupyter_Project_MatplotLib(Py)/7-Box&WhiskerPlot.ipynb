# Box Plot 

```python
import matplotlib.pyplot as plt
```

```python
# Box plot : use plt.boxplot function uses for stock or componies 's shares.
#                yellow line shows the median of the numbers .
# Whiskey plot : lower and upper part of the box plot is called whiskey plot.
# Outer layer : when we have the larger gap in the list, then it is denoted by "o" shape .

x =[10,20,30,40,50,60,70]
plt.boxplot(x)
plt.show()
```

# Parameters 

```python
# notch : change the shape of box it uses True and false as argument.
# vert  : uses for horizontal or vertical box plot. it also passes True or False as arguments.

# width : changing the width of the box
# labels : inside label pass a dict it uses for the changing the name of the x axis.

# patch_artist : use for filling colors inside the box.Passes True and False.
# showmeans    : uses for showing the mean of the plot.

# whis : it is used for joining the whiskey to the outler.
# sym  : for changing the color of the outer.and the shape of the outler.
#        give blank space for hiding the outler.

# boxprops : uses for changing the properties of the box and it passes dixtionary .

# capprops : we can change the properties of the cap and it also passes the dict as argument.

# whiskerprops : it changes the color of the whisk also passes dict as argument.

# filerprops : uses for changing filler properties.


x =[10,20,30,40,50,60,70,130]
plt.boxplot(x,notch=True,labels=["java"],patch_artist=True,showmeans=True,sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"),whiskerprops=dict(color="y"))
plt.show()
```

# Multiple box plot

```python
x =[10,20,30,40,50,60,70,130]
y =[10,30,20,40,50,90,40,150]

z = [x,y]

plt.boxplot(z,notch=True,labels=["python","java"],patch_artist=True,showmeans=True,sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"),whiskerprops=dict(color="y"))

plt.show()
```
