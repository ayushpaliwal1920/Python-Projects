```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")

df.info()
```

converted 0 and 1 value of seniorCitizen to yes/no  to make it easier to understand.


```python
def conv(value):
    if value == 1:
        return "Yes"
    else:
        return "No"
     
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
df.head(10)
```

```python
ax = sns.countplot(x="Churn",data=df)
ax.bar_label(ax.containers[0]) # for showing values.
plt.title("Count of customers by Churn")
plt.show() 
```

```python
# For showing percentage data.

gb = df.groupby("Churn").agg({"Churn":"count"})
gb
```

```python
plt.figure(figsize=(3,4))
plt.pie(gb['Churn'],labels=gb.index,shadow=True,colors=["r","g"],autopct="%1.2f%%")
plt.title("Percentage Of Customers By Churn")
plt.show()
```

From the given pie chart we can conclude that 26.54% of our customers have churned

```python
plt.figure(figsize=(4,4))
sns.countplot(x="gender",data= df,hue="Churn")
plt.title("Churn By Gender")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x="SeniorCitizen",data= df,hue="Churn")
plt.title("Churn By SeniorCitizen")
plt.show()
```

```python
counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Step 2: Convert to percentage
percentages = counts.div(counts.sum(axis=1), axis=0) * 100
plt.figure(figsize=(3,3))

# Step 3: Plot stacked bar chart
ax = percentages.plot(
    kind='bar',
    stacked=True,
    figsize=(5,5),
    color=['#1f77b4', '#ff7f0e']
)

# Step 4: Add percentage labels
for i, row in enumerate(percentages.values):
    cumulative = 0
    for j, val in enumerate(row):
        if val > 0:
            ax.text(i, cumulative + val / 2, f'{val:.1f}%', ha='center', va='center', color='white', fontsize=10)
        cumulative += val

# Step 5: Beautify
plt.title("Churn by SeniorCitizen (Percentage)")
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage (%)")
plt.legend(title="Churn", bbox_to_anchor = (0.9,0.9))
plt.ylim(0, 100)
plt.show()
```

```python
# List of columns
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaperlessBilling', 'PaymentMethod'
]

# Setup figure grid
plt.figure(figsize=(18, 20))

# Loop through columns and create subplots
for i, col in enumerate(cols, 1):
    plt.subplot(4, 3, i)  # 4 rows, 3 columns of plots
    sns.countplot(data=df, x=col, hue='Churn', palette='rocket_r')
    plt.title(f"Churn by {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=30, ha='right')  # Rotate x labels if needed

plt.tight_layout()
plt.show()
```

Customers with no internet-related services or short-term contracts tend to churn less, while those using month-to-month contracts, paperless billing, and multiple add-on services (like streaming or online backup) show higher churn rates.

```python
plt.figure(figsize=(8,4))
ax = sns.countplot(x="PaymentMethod",data=df,hue="Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churned Customers Payment method")
plt.xticks(rotation =45)
plt.show()
```
