# Scipy Cluster

Clustering is the task of dividing the population or data into a number of groups such that data points in the same group  are more similar to other data points in the same group and diddimilar to the data points in other group.

Each group which is formed from clustering is known as a cluster.

There are two types of clusters , which are : o.Central o.Hierarchy 

```python
# In the demo, we will use Numpy to generate two sets of random points.
# After joining both these sets , we whiten the data .
# Whitening normaslizes the data and is an essential step before using k means clustering.
# finally,we use the K-means functions and pass it the pass it the data and number of clustered we want.

import numpy as np
from scipy.cluster.vq import kmeans,whiten
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")

# Create 50 datapoints in two cluster a and b 

pts = 100

a = np.random.multivariate_normal([0,0],[[4,1],[1,4]],size=pts)
b = np.random.multivariate_normal([30,10],[[10,2],[2,1]],size=pts)
features = np.concatenate((a,b))

# whiten data :

whitened = whiten(features)

# find 2 cluster in the data :

codebook, distrortion = kmeans(whitened,2)

# plot whitened data and cluster centre in red :
plt.scatter(whitened[:, 0], whitened[:,1])
plt.scatter(codebook[:,0],codebook[:,1],c='r')
plt.show()



```
