# Encoding



Encoding means converting categorical (non-numeric) data into numeric format so that machine learning models can process it.

# One hot Encoding

uses for not labeled data

```python
import pandas as pd

dataset = pd.read_csv("loan.csv")
dataset.head(10)
```

```python
# first handling data : we are encoding gender and married column >

dataset.isnull().sum()
```

```python
dataset["Gender"].fillna(dataset["Gender"].mode()[0],inplace=True)

dataset["Married"].fillna(dataset["Married"].mode()[0],inplace=True)
dataset.head(10)
```

```python
en_data = dataset[["Gender","Married"]]
en_data
```

```python
# One hot encoding : using get_dummies() method

pd.get_dummies(en_data)
```

```python
# One hot encoding : using scikit learn

# drop = "first " => uses when data is exceeded , so it deletes the first column of encoded data.
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop="first")
ar = ohe.fit_transform(en_data).toarray() # it returns a sparse matrix ,.toarray() converts the sparse matrix into a normal dense NumPy array.
ar
# Sparse Matrix : it is a matrix in which maximum content is filled with 0 and 1.
```

```python
pd.DataFrame(ar,columns=["Gender_Male","Married_Yes"])
```
