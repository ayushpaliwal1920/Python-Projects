```python
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 10000],  # Paradise Biryani
    [2, 120000, 140000, 10000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 2000, 240000, 270000],  # Burger Point
    [5, 160000, 1000, 205000, 230000]   # Chai Point
])
print("==== Xomato Sales analysis ====")
print("\n Sales data Shape :", sales_data.shape)
print("\n Data of first three restaurant :\n", sales_data[:3])
print("\n All rows and columns in given range :\n", sales_data[:,1:]) # all rows (:) , columns from 1 to last (,1:)

```

```python
# Total sales per year :

print("\nYearly total :\n",np.sum(sales_data,axis=0))

yearly_toatal = np.sum(sales_data[:, 1:],axis=0)

print("\nYearly total :\n",yearly_toatal)

# Minimum sales per restaurant :

min_sales = np.min(sales_data[: ,1: ] , axis=1)
print("\nData as per minimun sales :\n" , min_sales)

# Maximum sales per year :

max_sales = np.max(sales_data[: ,1:] , axis=0)
print("\nData as per maximum sales per year :\n" , max_sales)

# Average sales per restaurent

avg_sales = np.mean(sales_data[:,1:],axis=1)
print("\nData as per Average sales per restaurent :\n" , avg_sales)
```

```python
# Cumalitive sales :a running total of a sequence of numbers, where each value in the resulting list is the sum of that number and all the preceding numbers in the original sequence

cum_sum = np.cumsum(sales_data[:,1:], axis=1)
print("\nComalitive sum :\n" , cum_sum)


plt.figure(figsize=(10,6))
plt.plot(np.mean(cum_sum,axis=0))
plt.title("Average cumulative sales accross all restaurant")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector addition :", vector1 + vector2)

print("Multiplication vector :",vector1*vector2)

print("\nDot Product :", np.dot(vector1,vector2))

angle = np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print("\nAngle btw two vectors :",angle)
```

```python
# Vectorized operation : jab ek hi operation sab k sath ho

restaurant_types = np.array(["Biryani","chinese","pizza","burger","cafe"])

vectorized_upper = np.vectorize(str.upper)

print("vectorized upper :",vectorized_upper(restaurant_types))
```
