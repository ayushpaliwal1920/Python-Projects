# Insert

```python
import pandas as pd

var = pd.DataFrame({"A": [1,2,3,4,5,6] ,"B": [7,8,9,1,2,3]})
var

# We will use insert function for inseting data. It takes 3 parameters index,position and value.
# Length of Data must be equal.

var.insert(1,"Py",var["A"])  # Duplicate copy of A

var.insert(2,"Py1",[11,12,13,14,15,16]) 

var
```

```python
# when we need limited amount of Data. using Slicing

var = pd.DataFrame({"A": [1,2,3,4,5,6] ,"B": [7,8,9,1,2,3]})

var["py12"] = var["A"][:3]
var
```

# DELETE

```python
var = pd.DataFrame({"A": [1,2,3,4,5,6] ,"B": [7,8,9,1,2,3],"C": [7,81,9,1,12,2]})

# Deleting using POP fxn :

var1 = var.pop("B")
var1
print(var)

# Using Del Fxn or keyword :

del var["A"]
var
```
