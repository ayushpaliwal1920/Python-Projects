# Merge 


```python
import pandas as pd

# Merge : Merge krne k liye ek part dono dataframe mai same hona chahiye jise id kehte hai.

#  Agr ek column k elements alag hai to common elements choose kiye jate hai./

var1 = pd.DataFrame({"Name": ["Ayuhs","dave","Sara","John"],"Age":[22,12,24,22]})
var2 = pd.DataFrame({"Name": ["Ayuhs","dave","Manisha","John"],"City":["Uk","Mp","Ap","Hp"]})



pd.merge(var2,var1,on="Name")

# How parameter : uses for reperesenting missing data  how = inner or left or right or outer

pd.merge(var1,var2,how="right") # right vale var ka all data.
pd.merge(var1,var2,how="outer") # for all data



# for checking which data is present or not : use "indicator"= True  parameter
#  
pd.merge(var1,var2,how="outer",indicator=True)


```

```python
#  if both columns are same  : then use 'left_index' = True , 'right_index'=True parameter to see all data .and we can give b=new name to these columns by 'suffixes' parameter 

var1 = pd.DataFrame({"Name": ["Ayuhs","dave","Sara","John"],"Age":[22,12,24,22]})
var2 = pd.DataFrame({"Name": ["Ayuhs","dave","Manisha","John"],"Age":["Uk","Mp","Ap","Hp"]})

pd.merge(var1,var2,left_index=True,right_index=True,suffixes=("name","id"))
```

# Concat


```python
# Concat uses for join two series or dataframe.

#  merge() is used for joining data frames based on a common column or index, concat() is used for concatenating data frames either vertically or horizontally.

# In cancat fxn we gives data in list []

sr1 = pd.Series([1,2,3,4,5])
sr2 = pd.Series([11,12,13,14,15])

pd.concat([sr1,sr2])
```

```python
d1 = pd.DataFrame({"Name": ["Ayuhs","dave","Sara","John"],"Age":[22,12,24,22]})
d2 = pd.DataFrame({"N": ["Ayuhs","John"],"City":["Uk","Hp"]})

# concat dataframes 

pd.concat([d1,d2])

# merge according to the axis .

pd.concat([d1,d2],axis=1) # axis = 1, merge according to the column.

# we can remoce NaN : use "join=outer or inner" parameter , outer shows union and inner shows intersection

pd.concat([d1,d2],axis=1,join="inner")
```

```python
# "key" parameter : uses for  making key for merging data


df1 = pd.DataFrame({"Name": ["Ayuhs","dave","Sara","John"],"Age":[22,12,24,22]})
df2 = pd.DataFrame({"Name": ["Ayuhs","dave","Manisha","John"],"City":["Uk","Mp","Ap","Hp"]})

pd.concat([d1,d2],axis=1,keys=["d1","d2"])
```
