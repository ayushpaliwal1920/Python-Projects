# DataFrames


```python
# For creating data frames we need lists or Dict :

import pandas as pd


# DataFrame using list

l = [1,2,2,3,5,5,5,5,4]
b = [1,2,22,3,5,5,5,5,4]



var = pd.DataFrame(l,b)
print(var)
print(type(var))
```

```python
# DataFrame using Dict :

# Imp = when we are passin data from dictionary the length of all data should be same or error will show

d = {"a":[1,2,4,53,2,4,4],"s":[11,22,44,55,66,77,88],"d":["Ayush",22,"dave",55,66,"Daemon",88],1:[11,"Abhi",44,55,66,77,88]}

var1 = pd.DataFrame(d)
print(var1)
print(type(d))

# functions or parameter for datafames :

# when we need a Particular column then we uses "column" attribute
# we can change index of the dataframe by using 'index=[]' parameter.
 

print("Particular column:")
var2 = pd.DataFrame(d,columns=["a","d",1],index=["a","b","c",'d',"e","f","g"])
print(var2)


# get a Particular value for DataFrame:  just like arrays
 
print(var1["a"][3])
print(var2["d"][2],var2[1][1])
```

```python
# Creating Dataframe form multiple lists =>

list_1 = [[11,22,33,44],[55,66,77,88],[101,102,103,104],[1.0,1.2,1.3,1.4]]

var3 = pd.DataFrame(list_1)
print(var3)

# Get element from dataFrame =>

print(list_1[1][3])
print(list_1[0][1])
```

```python
# Creating DataFrame from a Series =:

sr = {"ayush":pd.Series([1,2,3,4,5]),"Dave":pd.Series([11,22,33,44,55]),"Sara":pd.Series(['s','d','g','h','j'])}

var4 = pd.DataFrame(sr)
print(var4)
```
