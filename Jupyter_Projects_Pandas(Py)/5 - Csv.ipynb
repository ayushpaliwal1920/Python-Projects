#Creating CSV File

```python
# Creating and writing CSV file

# A CSV file is just tabular data (rows & columns).

# A Pandas DataFrame is a Python object that also represents tabular data.

# Since they both have the same structure, Pandas lets you easily convert a DataFrame â†’ CSV file.

import pandas as pd

dict = {"a":[1,2,3,4,5],"b":[4,5,6,7,8],"c":[9,1,0,2,3]}

d = pd.DataFrame(dict)
print(d)

# now create a csv file using to_csv("Name") fxnc:

# d.to_csv("Test_new.csv")

# creating a csv file without index passing argument index = false :

d.to_csv("Test_new.csv",index=False)

# Changing or modifying Head Data using header parameter :

d.to_csv("Test_new1.csv" , index=False , header=[1,2,3])
```
