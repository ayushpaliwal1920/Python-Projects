# Arithmetic Operations


```python
import pandas as pd

var = pd.DataFrame({"A": [1,2,3],"B":[4,5,6],"C": [7,8,9]})
var
```

```python
# Column-wise Operations :

var["D"] = var["A"] + var["B"]
var["E"] = var["A"] - var["B"]
var["F"] = var["A"] * var["B"]
var["G"] = var["A"] / var["B"]

var
```

```python
# Logical Operations :

var1 = pd.DataFrame({"A": [10,20,30],"B":[40,50,60],"C": [70,80,90]})

var1["PyA"] = var1["A"] <=20

var1["PyB"] = var1["A" and "B"] <= 30

var1["PyC"] = var1["A" or "B"] <= 30

var1
```

```python
# Element-wise Operations  :

var1 = pd.DataFrame({"A": [10,20,30],"B":[40,50,60],"C": [70,80,90]})

var1["Mul2"] = var1["A"]*2
var1["Mul3"] = var1["B"]*3.0
var1
```

```python
# Operations on Entire DataFrame  :

var1 = pd.DataFrame({"A": [10,20,30],"B":[40,50,60],"C": [70,80,90]})

var1 + 300
```

```python
# Between Two DataFrames / Series  :

var1 = pd.DataFrame({"A": [10,20,30],"B":[40,50,60],"C": [70,80,90]})
var3 = pd.DataFrame({"A": [100,200,300],"B":[400,500,600],"C": [700,800,900]})

var1 + var3

# Using Pandas Built-in Methods  :

# var1["A"].add(var3["B"], fill_value=0) 

var1["A"].mul(var1["B"], fill_value=0)
```
