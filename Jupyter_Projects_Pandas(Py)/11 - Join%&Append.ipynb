# JOIN

```python
import pandas as pd

#  Join fxn : used for joining two data frames , it works properly if we have unmatched data.

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a",'b','c','d'])
var2 = pd.DataFrame({"C":[10,20],"D":[11,22]},index=["a",'b'])

var1.join(var2)

# if we want to give coustomizd index we have to give it to both dataframes.

var2.join(var1)

# if we want to data according to ourself : use 'how' parameter

var2.join(var1,how="cross") # cartisian project
var2.join(var1,how="outer") # union
var2.join(var1,how="inner") # intersectionz
var2.join(var1,how="right") # according to var1
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a",'b','c','d'])
var2 = pd.DataFrame({"C":[10,20],"B":[11,22]},index=["a",'b'])

# if we have same column name then : use suffix parameter, through suffix name we can change column name during runtime

var1.join(var2,how="outer" , lsuffix="_1",rsuffix="_2") # for left(var1) use lsuffix and rsuffix for right one
```
