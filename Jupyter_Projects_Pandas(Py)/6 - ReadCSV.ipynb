# Read Csv File

```python
# for modifying or changing a large number of data which is in csv file we use pythonor panda.

import pandas as pd

# Read CSV => var = pd.read_csv("path\name.csv") 
# use double backslash at the starting and where sigle backslash is!

csv_1 = pd.read_csv("C:\\PythonProjects\\Jupyter_Projects\\students_large.csv")

print(csv_1)
```

```python
# Get rows to the particular indices :  using "nrows= " attribute.

csv_2 = pd.read_csv("C:\\PythonProjects\\Jupyter_Projects\\students_large.csv",nrows= 12)

print(csv_2)

#  Get Particular column : using "usecols=["Age","name"]  or usecols =[0,1,2]" attribute

csv_3 = pd.read_csv("C:\\PythonProjects\\Jupyter_Projects\\students_large.csv",nrows= 12, usecols= [1,2,3])

print(csv_3)

```

```python
# Skip any row : use skiprows=[] or skiprows = reange(5,25)

csv_4 = pd.read_csv("C:\\PythonProjects\\Jupyter_Projects\\students_large.csv" , skiprows=[0,1,2,3,4,5])
print(csv_4)
```

```python
# Change any column into index   : using Index_col

csv_5 = pd.read_csv("C:\\PythonProjects\\Jupyter_Projects\\students_large.csv" , index_col="City")
print(csv_5)

# Make header or change header : using parameter "header"= row number


print("header changed")
csv_6 = pd.read_csv("C:\\PythonProjects\\Jupyter_Projects\\students_large.csv"  , header=5 , usecols=[5] , skiprows= range(5,30))

print(csv_6)
```

```python
# Change name of header : using name parameter name = []

csv_5 = pd.read_csv("C:\\PythonProjects\\Jupyter_Projects\\students_large.csv" , names=["Col1","Col2","Col3","Col4","Col5","Col6","Col7","Col8"])
print(csv_5)
```

```python
# prefix parameter : error is showing ,That’s because in newer versions of Pandas (>=2.0), the prefix argument was removed from read_csv().

csv_6 = pd.read_csv("C:\\PythonProjects\\Jupyter_Projects\\students_large.csv" , header=None )

# another way 
csv_6.columns = [f"Col{i}" for i in range(csv_6.shape[1])]  

#csv_6.shape[0] → number of rows
#csv_6.shape[1] → number of columns

print(csv_6)
```

```python
# Make int to Float Data : using "dtype= []" parameter.

csv_1 = pd.read_csv("C:\\PythonProjects\\Jupyter_Projects\\students_large.csv", dtype={"Maths": "float" ,"Physics":"float","Chemistry":"float"})

print(csv_1)


```
